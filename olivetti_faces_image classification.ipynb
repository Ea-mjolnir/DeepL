```python
import os
from sklearn.datasets import fetch_olivetti_faces

if not os.path.exists('./datasets_olivetti_faces'):
    os.mkdir('./datasets_olivetti_faces')

X_faces,y_labels = fetch_olivetti_faces(data_home=r'./datasets_olivetti_faces',shuffle=False, random_state=0, download_if_missing=True, return_X_y=True)

print("Input :", X_faces.shape)
print("Labels:", y_labels.shape)
```

```python
n_samples, n_features = X_faces.shape
print(n_features)
```

```python
print(y_labels)
```

```python
import matplotlib.pyplot as plt
import numpy as np

n_row, n_col = 2, 3
n_components = n_row * n_col
image_shape = (64, 64)

def plot_faces(title, images, image_names, n_col, cmap=plt.cm.gray):
    n_images = len(images)
    n_row = int(np.ceil(n_images / n_col))
    n_to_display = min(n_images, 10)  # Display a maximum of 10 images
    fig, axs = plt.subplots(
        nrows=n_row,
        ncols=n_col,
        figsize=(1.5 * n_col, 2.0 * n_row),
        facecolor="white",
        constrained_layout=True,
    )
    fig.set_constrained_layout_pads(w_pad=0.01, h_pad=0.02, hspace=0, wspace=0)
    fig.set_edgecolor("black")
    fig.suptitle(title, size=16)
    
    for i, (ax, vec, name) in enumerate(zip(axs.flat, images[:n_to_display], image_names[:n_to_display])):
        vmax = max(vec.max(), -vec.min())
        ax.imshow(
            vec.reshape(image_shape),
            cmap=cmap,
            interpolation="nearest",
            vmin=-vmax,
            vmax=vmax,
        )
        ax.set_title(name)
        ax.axis("off")

    # Remove empty subplots (if any)
    for i in range(n_to_display, n_col * n_row):
        fig.delaxes(axs.flat[i])

    plt.show()

# Example usage:
n_col = 5
plot_faces("Faces from olivetti_faces", X_faces, y_labels, n_col)
```

```python
# Global centering (focus on one feature, centering all samples)
faces_centered = X_faces - X_faces.mean(axis=0)

# Local centering (focus on one sample, centering all features)
faces_centered -= faces_centered.mean(axis=1).reshape(n_samples, -1)


# Example usage:
n_col = 5
plot_faces("Faces from olivetti_faces", faces_centered, y_labels, n_col)
```

```python
from collections import Counter

y_labels

# Count the unique faces based on y_labels
unique_faces = Counter(y_labels)

# Get the count of unique faces
num_unique_faces = len(unique_faces)

print("Number of different faces:", num_unique_faces)
print("unique target number:",np.unique(y_labels))
```

```python
import matplotlib.pyplot as plt


def show_distinct_faces(images, unique_ids):
    
    image_shape = (64, 64)
    
    # Creating 4x10 subplots in 18x9 figure size
    fig, axarr = plt.subplots(nrows=4, ncols=10, figsize=(18, 9))
    
    # For easy iteration, flattened 4x10 subplots matrix to a 40-array
    axarr = axarr.flatten()
    
    # Iterating over user ids
    for unique_id in unique_ids:
        image_index = unique_id * 10  # Corrected indexing
        axarr[unique_id].imshow(images[image_index].reshape(image_shape), cmap='gray')
        axarr[unique_id].set_xticks([])
        axarr[unique_id].set_yticks([])
        axarr[unique_id].set_title("face id: {}".format(unique_id))
    
    plt.suptitle("There are {} distinct people in the dataset".format(len(unique_ids)))
    plt.show()

    
# Example usage:

unique_faces= np.unique(y_labels)
show_distinct_faces(X_faces, unique_faces)
```

```python
from sklearn.model_selection import train_test_split

X_training_faces, X_test_faces, y_training_labels, y_test_lables = train_test_split(X_faces, y_labels, test_size=0.2, random_state=0,shuffle=True)

print(f'The dataset comprises of {X_training_faces.shape[0]} samples.')
print(f'The test dataset comprises of {X_test_faces.shape[0]} samples.')
```

```python
X_train, X_valid, y_train, y_valid  = train_test_split(X_training_faces, y_training_labels, test_size=0.3, random_state=42,shuffle=True)


print(f'The training dataset comprises of {X_train.shape[0]} samples.')

print(f'The validation dataset comprises of {X_valid.shape[0]} samples.')
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense

model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(X_train.shape[1],)),
    tf.keras.layers.Dense(500, activation='relu'),
    tf.keras.layers.Dense(300, activation='relu'),
    tf.keras.layers.Dense(200, activation='relu'),  
    tf.keras.layers.Dense(100, activation='relu'),
    tf.keras.layers.Dense(40, activation='softmax')])

model.summary()
```

```python
print(type(model.weights))

print(len(model.weights))
print(model.weights[0])
```

```python
#Checking of weight size and biase size for the first layer 
combined_params = model.weights[0].numpy()
weights_size = combined_params[:].size
biases_size = combined_params[-1].size 

print("Size of the weights:", weights_size)
print("Size of the biases:", biases_size)

#Note thier summation gives the total parameter size for the given layer
para_size=weights_size + biases_size
print ("parameter_size:",para_size)
```

```python
#Training the model 
model.compile(optimizer =tf.keras.optimizers.Nadam(learning_rate=0.001),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy'])

callbacks_list = [
    tf.keras.callbacks.EarlyStopping(monitor= 'accuracy', min_delta = .0001, patience =10, restore_best_weights=True)
]

h = model.fit(X_train,y_train, batch_size=32,epochs = 50, validation_data = (X_valid, y_valid),callbacks = callbacks_list)
```

```python
# Loss is too high.
model.evaluate(X_test_faces, y_test_lables)
```

```python
import matplotlib.pyplot as plt

# Create a figure with two subplots side by side
fig, axs = plt.subplots(1, 2, figsize=(8, 3))

# Plotting the training and validation loss
axs[0].plot(h.history['loss'], label='Training Loss')
axs[0].plot(h.history['val_loss'], label='Validation Loss')
axs[0].set_title('Training and Validation Loss')
axs[0].set_xlabel('Epoch')
axs[0].set_ylabel('sparse_categorical_crossentropy')
axs[0].legend()

# Plotting the training and validation Accuracy.
axs[1].plot(h.history['accuracy'], label='Training_Accuracy')
axs[1].plot(h.history['val_accuracy'], label='Validation_Accuracy')
axs[1].set_title('Training and Validation Accuracy')
axs[1].set_xlabel('Epoch')
axs[1].set_ylabel('Accuracy')
axs[1].legend()

plt.tight_layout() 
plt.show()
```
